```python
import numpy as np
from numpy import random as rn
import scipy.stats as stat
import matplotlib.pyplot as plt
```

```python
eu = np.exp(1)
```

```python
def dist_rand(N):

    x = (1 - (1/eu**2)) * rn.random(N) + (1/eu**2)
    return -np.log(x)
    
```

```python
F = lambda t: np.exp(-t**2)     # We want to integrate this function
g = lambda t: np.exp(-t)        # Sampling disterbution 
```

```python
plt.hist(dist_rand(1000000), bins=100, density=True)
plt.xlabel("x")
plt.ylabel("Density")
plt.title("Exponential Distribution(Unnormalized)")
plt.savefig("img/P5N1.png", dpi = 300)
```

```python
def simple_estimate(M):

    samples = rn.random(M)*2
    f_bar = np.mean(F(samples))
    estimate = f_bar * 2

    return estimate
```

```python
simple_estimate(100000)
```

```python
def important_estimate(M):
    samples = dist_rand(M)
    f_bar = np.mean(F(samples) / g(samples))
    estimate = f_bar * (1 - np.exp(-2))

    return estimate
```

```python
important_estimate(100000)
```

```python
real_ans = 0.8820813907624216799674810359140540372240517708685564680103480660
n_list = range(100, 1000000, 1000)

plt.scatter(n_list, [simple_estimate(i) for i in n_list], s = 0.8, color = "blueviolet", label = "Simple Estimation")
plt.plot([1, 1010000], [real_ans, real_ans], linestyle = "dashed", color = "forestgreen", label = "Real Value")
plt.ylim([0.87, 0.90])
plt.xlabel("N")
plt.ylabel("Integral")
plt.title("Simple Method")
plt.legend();
plt.savefig("img/P5N2.png", dpi = 300)
```

```python
plt.scatter(n_list, [important_estimate(i) for i in n_list], s = 0.8, color = "blueviolet", label = "Important Estimation")
plt.plot([1, 1010000], [real_ans, real_ans], linestyle = "dashed", color = "forestgreen", label = "Real Value")
plt.ylim([0.87, 0.90])
plt.xlabel("N")
plt.ylabel("Integral")
plt.title("Important Method")
plt.legend();
plt.savefig("img/P5N3.png", dpi = 300)
```
