```python
import numpy as np
from numpy import random as rn
import scipy.stats as stat
from scipy.linalg import norm
import matplotlib.pyplot as plt
```

```python

ro_0 = 1.0
F = lambda x: x * ((ro_0 / 4) * x + (5*ro_0 / 4))
```

```python
def sphere_gen(N):
    M = int(N * 7.0 / np.pi)
    rand_points = 2 * rn.random((M, 3)) -  [1, 1, 1]
    sphere_points = rand_points[[norm(rand_points[j, :]) < 1.0 for j in range(M)]]
    return sphere_points[0:N, :]
```

```python
def calc_rx (samples):

    x_rn = samples[:, 0]
    drx_m = np.mean(F(x_rn)) * (4 * np.pi / 3) 
    return drx_m
```

```python
rn_points = sphere_gen(10000)
plt.scatter(rn_points[:, 0], rn_points[:, 1], s = 0.5)
plt.xlabel("x")
plt.ylabel("y")
plt.title("2D Projection of Uniform Sampling of Sphere")
plt.savefig("img/P6N1.png", dpi = 300)
```
