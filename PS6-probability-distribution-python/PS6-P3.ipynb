```python
import numpy as np
from numpy import random as rn
import scipy.stats as stat
import matplotlib.pyplot as plt
```

```python
def sum_rands(a, b, N):
    return np.sum(rn.randint(a, b, N))
```

```python
M = 100000
N = 5

sum_data = np.zeros(M)

for i in range(M):

    sum_data[i] = sum_rands(1, 10, N)


data_mean = np.mean(sum_data)
data_var = np.var(sum_data)
data_std = np.sqrt(data_var)



x = np.linspace(data_mean - 5*data_std, data_mean + 5*data_std, 1000)

plt.hist(sum_data, bins=20, rwidth=0.8, density=True, color = "teal")
plt.plot(x, stat.norm.pdf(x, data_mean, data_std), color = "darkred", lw = 2, linestyle = "dashed")
plt.title("Sum of N = {} numbers".format(N) + " with M = {} trials".format(M))
plt.xlabel("Sum of numbers")
plt.ylabel("Density")

plt.savefig("img/P3N1.png", dpi = 300)
```
