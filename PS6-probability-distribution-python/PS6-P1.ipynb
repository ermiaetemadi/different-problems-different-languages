```python
import numpy as np
from numpy import random as rn
import matplotlib.pyplot as plt
```

```python
rn_array = rn.randint(1, 11, 200000)

plt.hist(rn_array, bins=range(1, 12), rwidth=0.8, density=True, color = "teal")
plt.plot([1, 11], [0.1, 0.1], color = "darkred", lw = 1.5, linestyle = "dashed")
plt.title("Uniform Distribution (N = 200000)")
plt.xlabel("Number")
plt.ylabel("Density")
plt.savefig("img/P1N1.png", dpi = 300)
```

```python
stdn_list = np.zeros(200)
x = np.linspace(1, 200000, 200)

for n in range(1, 20000, 100):
    rn_array = rn.randint(1, 11, n)
    stdn_list[n//100] = np.std(np.histogram(rn_array)[0]) / n
    

plt.scatter(x, stdn_list, s = 2, color = "purple", label = "data std")
plt.plot(x, 1 / np.sqrt(x), color = "darkred", lw = 1.5, linestyle = "dashed", alpha = 0.7, label = "1 / sqrt(N)")
plt.ylim([0, .05])
plt.title("Standard Deveation Limit")
plt.ylabel("std")
plt.xlabel("N")
plt.legend()
plt.savefig("img/P1N2.png", dpi = 300)
```
