```python
import numpy as np
from numpy import random as rn
import scipy.stats as stat
import matplotlib.pyplot as plt
```

```python
N = 1000000
sigma = 2

x1 = rn.random(N)
x2 = rn.random(N)

F1 = lambda t: np.sqrt(-2*(sigma**2)*np.log(t))
F2 = lambda t: 2*np.pi*t

ro = F1(x1)
theta = F2(x2)

y1 = ro*np.sin(theta)
y2 = ro*np.cos(theta)
```

```python
p_x = np.linspace(-5*sigma, 5*sigma, 10000)
plt.hist(y1, bins=200, density =  True)
plt.plot(p_x, stat.norm.pdf(p_x, 0, sigma), color = "darkred", lw = 2, linestyle = "dashed")
plt.title("Normal Distribution")
plt.xlabel("x")
plt.ylabel("Density")
plt.savefig("img/P4N1.png", dpi = 300)
```
