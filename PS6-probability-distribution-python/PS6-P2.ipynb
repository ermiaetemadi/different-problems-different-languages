```python
import numpy as np
from numpy import random as rn
import matplotlib.pyplot as plt
```

```python
rn_array = rn.randint(1, 11, 20)

rn_array
```

```python
rn_array[1:][(rn_array[:] == 4.0)[:-1]]
```

```python
rn_array = rn.randint(1, 11, 2000000)

col_rn = rn_array[1:][(rn_array[:] == 4.0)[:-1]]

plt.hist(col_rn, bins=range(1, 12), rwidth=0.8, density=True, color = "teal")
plt.plot([1, 11], [0.1, 0.1], color = "darkred", lw = 2, linestyle = "dashed")

plt.title("Numbers After 4 Distribution (N = 2000000)")
plt.xlabel("Number")
plt.ylabel("Density")
plt.savefig("img/P2N1.png", dpi = 300)
```
